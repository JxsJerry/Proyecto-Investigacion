```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("Datos - Industria musical.xlsx", header = 0)
data
```

```python
plt.figure(figsize = (15,8))
plt.bar(data.Año, data.Ingreso)
plt.xlabel("Año")
plt.ylabel('Ingresos (Miles de millones de dólares)')
plt.title("Ingresos por venta de música física en la industria musical a nivel mundial 2001-2020")
plt.xticks([2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020],
           ["2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010",
            "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"])
plt.show()
```
